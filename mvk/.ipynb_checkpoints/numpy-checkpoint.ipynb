### Neden Numpy ?
* Veri saklama açısından listelere kıyasla daha optimize ve tek bir tipte eleman tutabilme

```python
a = [1,2,3,4]
b = [2,3,4,5]

ab = []
for i in range(len(a)):
    ab.append(a[i]*b[i])
ab
```

```python
a = np.array([1,2,3,4])
b = np.array([2,3,4,5])
a*b
```

### Array Birleştirme (concatenation) 

```python
import numpy as np
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
z = np.array([7,8,9])
```

```python
np.concatenate([x,y,z])
```

```python
np.hstack((x,y,z))
```

```python
np.vstack((x,z,y))
```

#### İki boyut

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,9]])
c = np.array([[10,11]])
```

```python
np.concatenate([a,b] , axis = 0)
```

```python
d = np.concatenate([a,c.T], axis = 1)
d
```

#### Axis Kavramı

```python
d.max(axis=1)
```

```python
d.min(axis=0)
```

```python
d.sum(axis=0)
```

### Array Ayırma (splitting)

```python
import numpy as np
```

```python
x = np.array([1,2,3,99,99,3,2,1])
```

```python
array1,array2,array3 = np.split(x,[3,5]) # 3 e kadar , 5 e kadar böl
```

```python
array1
```

#### İki boyut 

```python
m = np.arange(16).reshape(4,4)
m
```

##### Vsplit

```python
np.vsplit(m,[2])
```

```python
üst,alt = np.vsplit(m,[2])
```

```python
np.concatenate([üst,alt],axis=1)
```

##### Hsplit

```python
np.hsplit(m,[2])
```

```python
sol,sağ = np.hsplit(m,[2])
```

```python
np.concatenate([sol,sağ],axis=0)
```

### Fancy Index ile Elemanlara Erişmek ( Önemli ! ) 

```python
import numpy as np
```

```python
v = np.arange(0,30,3)
v
```

```python
v[[1,3,5]] # 1,3 ve 5.index deki değerleri getir.
```

#### İki boyut

```python
m = np.arange(9).reshape(3,3)
m
```

```python
m[[0,1],[2,0]] # 0.satır 2.sütundaki değer ile 1.satır 0.sütundaki değeri getir. (capraz çalışıyor dikkat!)
```

##### Index ile fancy index Kullanımı

```python
m[0,[1,2]] # 0.satırın 1.ve 2.sütununu getir
```

##### Slice ile fancy index kullanımı

```python
m[1:,[1,2]]
```

### Matematiksel İşlemler

```python
v / 5
np.divide(v,5)
```

```python
v * 5 / 10 - 1
function = lambda x : np.subtract(np.divide(np.multiply(x,5),10),1)
function(v)
```

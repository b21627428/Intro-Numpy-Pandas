## Kmeans Clustering 
* ki adet centroid belirlenir. Pointler hangi centroid e yaknsa kmelenip araya izgi ekilir.Kmelerin ortalamas alnp yeni centrodler belrilenir.Ne zamanki ekilen izgi ayn kalyorsa algoritma sonlanr.



* K deerini belirleme iin WCSS metrii kullanlr.
    * WCSS = within cluster sum of squares (ne kadar az olursa daha iyi.)
    * K ile WCSS arasnda tradeoff vardr.100 data varsa n i 100 seersek WCSS deeri 0 gelir.Fakat 100 adet kmeleme mantkl deildir model yava alaca iin.
    

### Implementation 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Dataset oluturma
#class1
x1 = np.random.normal(25,5,(1000))
y1 = np.random.normal(25,5,(1000))
#class2
x2 = np.random.normal(55,5,(1000))
y2 = np.random.normal(60,5,(1000))
#class3
x3 = np.random.normal(65,5,(1000))
y3 = np.random.normal(15,5,(1000))
```

```python
x = np.concatenate((x1,x2,x3),axis=0)
y = np.concatenate((y1,y2,y3),axis=0)
```

```python
df = pd.DataFrame({"x":x,"y":y})
```

```python
df.head()
```

```python
plt.scatter(x1,y1,color="black")
plt.scatter(x2,y2,color="black")
plt.scatter(x3,y3,color="black")
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,15),wcss)
plt.xlabel("k value")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(df)
```

```python
clusters
```

```python
df["label"] = clusters
```

```python
df
```

```python
df.label == 0
```

```python
df.x
```

```python
kmeans.cluster_centers_
```

```python
for i in range(3):
    plt.scatter(df.x[df.label == i],df.y[df.label == i])
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="yellow")  # centroidler
plt.show()
```

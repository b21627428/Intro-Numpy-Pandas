# Naive Bayes 
* P(Math|X) = P(X|Math).P(Math) / P(X)
    * P(X|Math) = Likelihood
    * P(Math) = Prior probability
    * P(X) = Marginal Likelihood
    
    
    * P(Math) = 6/11
    * P(X) = 4/11 : Setiimiz alandaki oran
    * P(X|Math) = 3/6 : Setiimiz alandaki matematik hocasnn tm hocalara blm
    * P(Math|X) = 3/4 = %75 orannda seili siyah nokta matematik hocasdr.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("data.csv",sep=",")
```

```python
data.head()
```

```python
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
```

```python
data["diagnosis"] = [1 if i == "M" else 0 for i in data["diagnosis"].values]
```

```python
y = data["diagnosis"].values
x_data = data.drop(["diagnosis"],axis=1)
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
#train ve test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
print("accuracy of naive bayes: {}".format(nb.score(x_test,y_test)))
```

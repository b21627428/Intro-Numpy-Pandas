### Concantenation ( Array Birleştirme )  

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
```

```python
np.concatenate([x,y])
```

```python
z = np.array([7,8,9])
```

```python
np.concatenate([x,y,z])
```

```python
# iki boyut
a = np.array([[1,2,3],
              [4,5,6]])
a
```

```python
np.concatenate([a,a]) #axis = 0
```

```python
np.concatenate([a,a],axis=1)
```

```python
# iki boyut
m = np.arange(16).reshape((4,4))
m
```

```python
a,b = np.vsplit(m,[2])
np.concatenate([a,b],axis=1)
```

```python
a,b = np.hsplit(m,[2])
np.concatenate([a,b])
```

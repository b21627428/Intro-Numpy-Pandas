# İleri Toplulaştırma İşlemleri (Aggreagate ,filter ,transform ,apply )

```python
df = pd.DataFrame({
    'gruplar':["A","B","C","A","B","C"],
    'degisken1':[10,23,33,22,11,99],
    'degisken2':[100,253,333,262,111,969],
},columns=["gruplar","degisken1","degisken2"]
)
df
```

```python
# aggregate
```

```python
df.groupby("gruplar").describe()
```

```python
df.groupby("gruplar").aggregate(["min",np.median,max])
```

```python
df.groupby("gruplar").aggregate({"degisken1":["min","max"],"degisken2":[np.mean]})
```

```python
df
```

### Filter 

```python
def filter_func(x):
    return x["degisken1"].std()>9
```

```python
df.groupby("gruplar").std()
```

```python
df.groupby("gruplar").filter(filter_func)
```

### Transform

```python
df_a = df.iloc[:,1:]
df_a
```

```python
df_a.transform(lambda x: (x-x.mean()) / x.std() ) # sütün sütün çalışıyor.
```

### APPLY FUNCTION 

```python
df_a.apply(np.sum) #sütun bazlı çalışır.
```

```python
df_a.apply(lambda x: x-5)
```

```python
df_a
```

```python
df.groupby("gruplar").apply(np.sum)
```

### Pivot Tablolar

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
#pivot ile table
```

```python
titanic.pivot_table("survived",index="sex",columns="class")
```

```python
titanic.age.head()
```

```python
age = pd.cut(titanic["age"],[0,18,90]) #Sürekli değeri kategorilendirme
age.head(10)
```

```python
titanic.pivot_table("survived",["sex",age],"class").unstack()
```
